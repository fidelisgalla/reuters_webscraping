## This code is used to get all the stock in LQ45 for GARP analysis

This code will scrape the reuters financial highlights web page to get all financial data to conduct the calculation. This code is used to get all the EPS for 1 year, EPS for 5 year and the P/E ratio. This code will then calculate the PEG (PE Ratio/PER)for all the LQ45 stocks. From the list, an investor can take conclusion for choosing best stock based on the GARP analysis.

The list of LQ45 has been previously got from the previous code.

Creating function for all parameters needed. The function is get_soup : to get parsed soup element, get_price : get the latest price, get_time : get the current time, get_pe_now : get the current PE ratio, get_eps_1_year: get the last 1 year EPS, get_eps_3_year: get the last 3 year EPS, get_eps_5_year: get the last 5 year of EPS. 

NOTE: not all the function will be used

```python
def get_soup():
    r=requests.get(url_list)
    soup=BeautifulSoup(r.text,'html.parser')
    return soup

def get_price():
    stock=get_soup().find_all('span',attrs={'class':""})[0]  #function in function
    stock=float(stock.text.strip().replace(',',''))          
    return stock

def get_time():
    time=soup.find('span',attrs={'class':'nasdaqChangeTime'}).text.strip()
    return time

def get_pe_now():
    pe=get_soup().find_all('tr',attrs={'class':'stripe'})[4]
    pe=pe.find_all('td',attrs={'class':'data'})[0].text.strip()
    return pe

def get_eps_1_year():
    eps_one_year=get_soup().find_all('tbody',attrs={'class':'dataSmall'})
    eps_one_year=eps_one_year[0].find_all('td')[5].text.strip()
    return eps_one_year

def get_eps_3_year():
    eps_three_year=get_soup().find_all('tbody',attrs={'class':'dataSmall'})
    eps_three_year=eps_three_year[0].find_all('td')[6].text.strip()
    return eps_three_year

def get_eps_5_year():
    eps_five_year=get_soup().find_all('tbody',attrs={'class':'dataSmall'})
    eps_five_year=eps_five_year[0].find_all('td')[7].text.strip()
    return eps_five_year
```

```python
df_financial=pd.DataFrame({'Stock Price':price_list,'PER':per,'EPS Growth 1 Yr':eps_1,'EPS Growth 3 Yr':eps_3,'EPS Growth 5 Year':eps_5})
```
